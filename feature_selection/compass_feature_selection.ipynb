# COMPAS data feature selection

In this notebook we will briefly run few feature selection algorithms on two COMPAS datasets - raw and cleaned.

```python
import numpy as np
import pandas as pd
import os
from feature_selector import FeatureSelector
```

## Loading the data

```python
work_dir = os.getcwd()
print("Working directory %s" % work_dir)

data_dir = os.path.join(work_dir, 'data')
print("Data directory %s" % data_dir)
```

```python
compas_data_raw = pd.read_csv(os.path.join(data_dir, 'compas-scores-raw.csv'))
compas_data_cleaned = pd.read_csv(os.path.join(data_dir, 'compass_scores_cleaned.csv'))
```

## Feature selection of raw COMPAS data

```python
compas_data_raw.head()
```

```python
# Get the predicted labels

compas_labels = compas_data_raw['RawScore']
compas_data_raw = compas_data_raw.drop(columns = 'RawScore')
compas_labels.head(5)
```

```python
# Initialize feature selector 

fs = FeatureSelector(data = compas_data_raw, labels = compas_labels)
num_features = compas_data_raw.shape[1]
```

```python
# Constants for FeatureSelector

par_missing_threshold = 0.5
par_correlation_threshold = 0.5
```

The FeatureSelector has five functions for identifying columns to remove:

- identify_missing
- identify_single_unique
- identify_collinear
- identify_zero_importance
- identify_low_importance

These methods find the features to drop according to specified criteria. The identified features are stored in the ops attribute (a Python dictionary) of the FeatureSelector. We can remove the identified features manually or use the remove function in the FeatureSelector for actually removing the features.

## 1. Missing values

FeatureSelector provides a sufficient method to identify missing columns. The only parameter that is specified is *missing_threshold*.

FeatureSelector finds any columns with a missing fraction greater than specified threshold.

```python
fs.identify_missing(missing_threshold = par_missing_threshold)
```

We can see, that there is one column that misses at least 50% of the data. Therefore, such columns should be removed, if appropriate, or the data should be imputed.

```python
missing_columns = fs.ops['missing']
print("Missing columns:\n", missing_columns)
```

```python
fs.missing_stats.head(10)
```

Most of the data is fine, except of the **MiddleName** column which missess more than 74% of the data.

We could also plot the histograms of missing values.

```python
fs.plot_missing()
```

## 2. Single unique values

FeatureSelector also provides a convenient method to detect features(columns) that contain only one single unique value.

```python
fs.identify_single_unique()
```

```python
single_unique_features = fs.ops['single_unique']
print("Columns with just one unique value:\n", single_unique_features)
```

Inspecting the data we can see that **IsCompleted** contain only ones, whereas **IsDeleted** contains only zeros.
**AssessmentReason** contains only values Intake.

```python
fs.unique_stats.sample(num_features)
```

## Collinear features

This method finds pairs of collinear features based on the Pearson correlation coefficient. For each pair above the specified threshold (in terms of absolute value), it identifies one of the variables to be removed. We need to pass in a *correlation_threshold.*

For each pair, the feature that will be removed is the one that comes last in terms of the column ordering in the dataframe. (This method does not one-hot encode the data beforehand unless one_hot=True. Therefore correlations are only calculated between numeric columns)

Therefore, it would be convenient to hot-encode the data before calculating correlations, since we have a lot of text data.

```python
# Map the nationality column

nat_dict = {'African-American': 0, 'Arabic': 1, 'Asian': 2, 
            'Caucasian':3, 'Hispanic': 4, 'Native American': 5, 
            'Oriental': 6, 'Other': 7, 'African-Am': 0}

compas_data_raw['Ethnic_Code_Text'] = compas_data_raw['Ethnic_Code_Text'].map(nat_dict)
```

```python
compas_data_raw.head()
```

```python
# Map gender data

gen_dict = {'Male': 0, 'Female': 1}

compas_data_raw['Sex_Code_Text'] = compas_data_raw['Sex_Code_Text'].map(gen_dict)
```

```python
compas_data_raw.head()
```

```python
# Map legal status

leg_dict = {'Conditional Release': 0, 'Deferred Sentencing': 1, 'Parole Violator': 2, 
           'Post Sentence': 3, 'Pretrial': 4, 'Probation Violator': 5, 'Other': 6}

compas_data_raw['LegalStatus'] = compas_data_raw['LegalStatus'].map(leg_dict)
```

```python
# Map marital status

mar_dict = {'Divorced': 0, 'Married': 1, 'Separated': 2, 'Significant Other': 3,
           'Single': 4, 'Unknown': 5, 'Widowed': 6}

compas_data_raw['MaritalStatus'] = compas_data_raw['MaritalStatus'].map(mar_dict)
```

```python
# Map supervision levels

sup_dict = {'Low': 0, 'Medium': 1, 'Medium with Override Consideration': 2, 'High': 3}

compas_data_raw['RecSupervisionLevel'] = compas_data_raw['RecSupervisionLevel'].map(sup_dict)
```

After doing some category data encoding we can proceed with correlation computations.

```python
fs.identify_collinear(correlation_threshold = par_correlation_threshold)
```

We can see that correlation between features, by computing, is pretty low. Therefore, we must double-check the correlation by computing by hand.
